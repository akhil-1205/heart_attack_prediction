```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import copy
import math
```

```python
# Read the CSV file into a pandas DataFrame
df = pd.read_csv('train.csv')
```

```python
X_train = df.drop(columns = ['id', 'smoking_status', 'stroke'])
print(X_train)

Y_train = df['stroke']
```

One Hot Encoding

```python
#List of columns to encode
columns_to_encode = ['gender', 'ever_married', 'work_type', 'Residence_type']


X_train = pd.get_dummies(X_train, columns = columns_to_encode)

X_train
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression

#loading data
X = X_train
y = Y_train

# Create a logistic regression model
model = LogisticRegression()

# Fit the model to the data
model.fit(X, y)

# Make predictions
predictions = model.predict(X)

# Evaluate the model
accuracy = model.score(X, y)

print('Accuracy:', accuracy)
```

```python
df = pd.read_csv('test.csv')
#X_test = df.drop('smoking_status', axis = 1)
X_test = df.drop(columns = ['id', 'smoking_status'])
X_test = pd.get_dummies(X_test, columns = columns_to_encode)

X_test
```

```python
# Make predictions
predictions = model.predict(X_test)

probabilities = model.predict_proba(X_test)[:,1]


print(probabilities)
```
